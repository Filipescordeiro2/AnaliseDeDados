# Função de salvar

```python
def salvar_resultados_em_excel(df, caminho_arquivo):
    try:
        # Verificando se o diretório existe, caso não exista, criá-lo
        if not os.path.exists(os.path.dirname(caminho_arquivo)):
            os.makedirs(os.path.dirname(caminho_arquivo))

        # Salvando o DataFrame em um arquivo Excel
        df.to_excel(caminho_arquivo, index=False)
        print(f"Salvo em {os.path.dirname(caminho_arquivo)}")
    except Exception as e:
        print(f"Não foi possível salvar o arquivo: {e}")
```

# Calculos

# Analise - Album por Ano de 2014 a 2024

```python
media_albuns_por_ano = df_principal[df_principal['Ano de Lançamento'].between(2014, 2024)].groupby('Ano de Lançamento')['Nome do Álbum'].count()

# Calculando a média
media = round(media_albuns_por_ano.mean(),2)
print(f"Média: {media}")

# Calculando a moda
moda = media_albuns_por_ano.mode()[0]
print(f"Moda: {moda}")

# Calculando a mediana
mediana = media_albuns_por_ano.median()
print(f"Mediana: {mediana}")

# Calculando a amplitude
amplitude = media_albuns_por_ano.max() - media_albuns_por_ano.min()
print(f"Amplitude: {amplitude}")

# Calculando Q1 e Q3
Q1 = media_albuns_por_ano.quantile(0.25)
Q3 = media_albuns_por_ano.quantile(0.75)
print(f"Q1: {Q1}, Q3: {Q3}")

# Calculando o desvio padrão
desvio_padrao = round(media_albuns_por_ano.std(), 2)
print(f"Desvio Padrão: {desvio_padrao}")

# Calculando a assimetria
assimetria = round(media_albuns_por_ano.skew(), 2)
print(f"Assimetria: {assimetria}")

# Calculando o coeficiente de dispersão
coeficiente_dispersao = round(desvio_padrao / media, 2)
print(f"Coeficiente de Dispersão: {coeficiente_dispersao}")

# Interpretação dos resultados
print("Interpretação:")
print("A média de álbuns por ano é de aproximadamente 396.36 álbuns.")
print("A moda é de 70 álbuns, o que sugere que 70 foi o número mais frequente de álbuns lançados em um ano.")
print("A mediana é de 296 álbuns, o que indica que metade dos anos tiveram menos de 296 álbuns lançados.")
print("A amplitude é de 1088 álbuns, o que reflete uma grande variação no número de álbuns lançados entre os anos.")
print("O Q1 e Q3 sugerem que a maior parte dos anos teve entre 168 e 550.5 álbuns lançados.")
print("O desvio padrão é de aproximadamente 330.91 álbuns, o que indica uma variação considerável no número de álbuns lançados.")
print("A assimetria é de aproximadamente 1.35, o que sugere que a distribuição é assimétrica à direita, com mais anos tendendo a ter menos álbuns lançados e alguns anos tendendo a ter números maiores.")
print("O coeficiente de dispersão é de aproximadamente 0.83, o que indica uma dispersão moderada em relação à média, o que é coerente com o desvio padrão elevado e a assimetria à direita.")

# Criando uma tabela para mostrar os resultados
resultados = pd.DataFrame({
    'Média': [media],
    'Moda': [moda],
    'Mediana': [mediana],
    'Amplitude': [amplitude],
    'Q1': [Q1],
    'Q3': [Q3],
    'Desvio Padrão': [desvio_padrao],
    'Assimetria': [assimetria],
    'Coeficiente de Dispersão': [coeficiente_dispersao]
})

# Chamando a função para salvar os resultados em um arquivo Excel
salvar_resultados_em_excel(resultados,'C:\\Users\\USER\\Desktop\\Projeto_Senac\\Data_Result\\Analise_Album por Ano de 2014 a 2024\\Analise_Album por Ano de 2014 a 2024.xlsx')
```

# Analise - Reproducao do Spotify em 2023

# Analise Artista por Album de 2014 a 2024

```python
# Filtrando apenas álbuns lançados de 2014 a 2024
df_filtrado = df_principal[(df_principal['Ano de Lançamento'] >= 2014) & (df_principal['Ano de Lançamento'] <= 2024)]

# Calculando a média de álbuns por artista
media_albums = round(df_filtrado['Nome do Álbum'].value_counts().mean(), 2)
print(f"Média de Álbuns por Artista: {media_albums}")

# Calculando a moda de álbuns por artista
moda_albums = df_filtrado['Nome do Álbum'].mode()[0]
print(f"Moda de Álbuns por Artista: {moda_albums}")

# Calculando a mediana de álbuns por artista
mediana_albums = df_filtrado['Nome do Álbum'].value_counts().median()
print(f"Mediana de Álbuns por Artista: {mediana_albums}")

# Calculando a amplitude de álbuns por artista
amplitude_albums = df_filtrado['Nome do Álbum'].value_counts().max() - df_filtrado['Nome do Álbum'].value_counts().min()
print(f"Amplitude de Álbuns por Artista: {amplitude_albums}")

# Calculando Q1 e Q3 de álbuns por artista
Q1_albums = df_filtrado['Nome do Álbum'].value_counts().quantile(0.25)
Q3_albums = df_filtrado['Nome do Álbum'].value_counts().quantile(0.75)
print(f"Q1 de Álbuns por Artista: {Q1_albums}, Q3 de Álbuns por Artista: {Q3_albums}")

# Calculando o desvio padrão de álbuns por artista
desvio_padrao_albums = df_filtrado['Nome do Álbum'].value_counts().std()
print(f"Desvio Padrão de Álbuns por Artista: {desvio_padrao_albums}")

# Calculando a assimetria de álbuns por artista
assimetria_albums = df_filtrado['Nome do Álbum'].value_counts().skew()
print(f"Assimetria de Álbuns por Artista: {assimetria_albums}")

# Calculando o coeficiente de dispersão de álbuns por artista
coeficiente_dispersao_albums = desvio_padrao_albums / media_albums
print(f"Coeficiente de Dispersão de Álbuns por Artista: {coeficiente_dispersao_albums}")

# Encontrando o artista com a maior média de álbuns
artista_com_maior_media = df_filtrado.groupby('Artista')['Nome do Álbum'].count().idxmax()
print(f"Artista com a Maior Média de Álbuns: {artista_com_maior_media}")


# Análise Estatística:
print("Análise Estatística:")
print(f"Média de Álbuns por Artista: {media_albums} álbuns.")
print(f"Interpretação Estatística: A média representa o número médio de álbuns por artista. Em 2023, cada artista lançou, em média, {media_albums} álbuns, o que sugere que a maioria dos artistas lançou apenas um álbum, mas alguns lançaram mais, aumentando ligeiramente a média.")
print(f"Moda de Álbuns por Artista: '{moda_albums}'.")
print(f"Interpretação Estatística: A moda é o valor que mais apareceu entre os dados. No caso dos álbuns, '{moda_albums}' foi o álbum mais comum ou mais mencionado, destacando-se entre os outros lançamentos.")
print(f"Mediana de Álbuns por Artista: {mediana_albums} álbum.")
print(f"Interpretação Estatística: A mediana é o ponto central dos dados, o que significa que 50% dos artistas lançaram {mediana_albums} álbum ou menos, enquanto os outros 50% lançaram mais de {mediana_albums} álbum. A mediana ser igual a {mediana_albums} confirma que a maioria dos artistas lançou apenas um álbum.")
print(f"Amplitude de Álbuns por Artista: {amplitude_albums} álbuns.")
print(f"Interpretação Estatística: A amplitude, que é a diferença entre o maior e o menor número de álbuns lançados por um artista, é de {amplitude_albums} álbuns. Isso mostra uma enorme variação, com alguns artistas lançando apenas um álbum e outros lançando até {amplitude_albums}.")
print(f"Quartis (Q1 = {Q1_albums}, Q3 = {Q3_albums}):")
print(f"Interpretação Estatística: Tanto o Q1 quanto o Q3 são iguais a {Q1_albums}, indicando que 50% dos artistas (entre o primeiro e o terceiro quartil) lançaram exatamente {Q1_albums} álbum. A maioria dos artistas está bastante concentrada em torno desse valor, com uma distribuição bastante homogênea.")
print(f"Desvio Padrão de Álbuns por Artista: {desvio_padrao_albums} álbuns.")
print(f"Interpretação Estatística: O desvio padrão relativamente baixo indica uma pequena variação entre os artistas em termos do número de álbuns lançados. Isso sugere que a maioria dos artistas não se desvia muito do número de {Q1_albums} álbum.")
print(f"Assimetria de Álbuns por Artista: {assimetria_albums}.")
print(f"Interpretação Estatística: A assimetria extremamente alta revela que a distribuição é altamente inclinada à direita. Isso significa que a maioria dos artistas lançou poucos álbuns, enquanto poucos artistas lançaram muitos, elevando drasticamente a assimetria.")
print(f"Coeficiente de Dispersão de Álbuns por Artista: {coeficiente_dispersao_albums}.")
print(f"Interpretação Estatística: O coeficiente de dispersão mede a variação relativa em relação à média. Um valor de {coeficiente_dispersao_albums} sugere uma variação moderada no número de álbuns lançados, mas a média baixa e a alta assimetria indicam que a variação é impulsionada por um pequeno número de artistas que lançaram muitos álbuns.")


# Criando uma tabela para mostrar os resultados
tabela_resultados = pd.DataFrame({
    'Média de Álbuns por Artista': [media_albums],
    'Moda de Álbuns por Artista': [moda_albums],
    'Mediana de Álbuns por Artista': [mediana_albums],
    'Amplitude de Álbuns por Artista': [amplitude_albums],
    'Q1 de Álbuns por Artista': [Q1_albums],
    'Q3 de Álbuns por Artista': [Q3_albums],
    'Desvio Padrão de Álbuns por Artista': [desvio_padrao_albums],
    'Assimetria de Álbuns por Artista': [assimetria_albums],
    'Coeficiente de Dispersão de Álbuns por Artista': [coeficiente_dispersao_albums],
    'Artista com a Maior Média de Álbuns': [artista_com_maior_media]
})

salvar_resultados_em_excel(tabela_resultados,r'C:\Users\USER\Desktop\Projeto_Senac\Data_Result\Analise Artista por Album de 2014 a 2024\Analise Artista por Album de 2014 a 2024.xlsx')
```

# Analise - Musicas Repitidas por Artista de 2014 a 2024

```python
# Analisando artistas com músicas repetidas, filtrando por lançamentos de 2014 a 2024
df_filtrado_por_ano = df_filtrado[df_filtrado['Ano de Lançamento'].between(2014, 2024)]
musicas_repetidas = df_filtrado_por_ano['Nome da Música'].value_counts()
artistas_com_musicas_repetidas = df_filtrado_por_ano[df_filtrado_por_ano['Nome da Música'].isin(musicas_repetidas[musicas_repetidas > 1].index)]

# Calculando a quantidade de músicas repetidas por artista
quantidade_musicas_repetidas_por_artista = artistas_com_musicas_repetidas.groupby('Artista')['Nome da Música'].count()

# Calculando a média de músicas repetidas por artista
media_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.mean()
print(f"Média de Músicas Repetidas por Artista: {media_musicas_repetidas_por_artista:.2f}")

# Calculando a moda de músicas repetidas por artista
moda_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.mode()[0]
print(f"Moda de Músicas Repetidas por Artista: {moda_musicas_repetidas_por_artista}")

# Calculando a mediana de músicas repetidas por artista
mediana_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.median()
print(f"Mediana de Músicas Repetidas por Artista: {mediana_musicas_repetidas_por_artista:.2f}")

# Calculando a amplitude de músicas repetidas por artista
amplitude_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.max() - quantidade_musicas_repetidas_por_artista.min()
print(f"Amplitude de Músicas Repetidas por Artista: {amplitude_musicas_repetidas_por_artista:.2f}")

# Calculando Q1 e Q3 de músicas repetidas por artista
Q1_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.quantile(0.25)
Q3_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.quantile(0.75)
print(f"Q1 de Músicas Repetidas por Artista: {Q1_musicas_repetidas_por_artista:.2f}, Q3 de Músicas Repetidas por Artista: {Q3_musicas_repetidas_por_artista:.2f}")

# Calculando o desvio padrão de músicas repetidas por artista
desvio_padrao_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.std()
print(f"Desvio Padrão de Músicas Repetidas por Artista: {desvio_padrao_musicas_repetidas_por_artista:.2f}")

# Calculando a assimetria de músicas repetidas por artista
assimetria_musicas_repetidas_por_artista = quantidade_musicas_repetidas_por_artista.skew()
print(f"Assimetria de Músicas Repetidas por Artista: {assimetria_musicas_repetidas_por_artista:.2f}")

# Calculando o coeficiente de dispersão de músicas repetidas por artista
coeficiente_dispersao_musicas_repetidas_por_artista = desvio_padrao_musicas_repetidas_por_artista / media_musicas_repetidas_por_artista
print(f"Coeficiente de Dispersão de Músicas Repetidas por Artista: {coeficiente_dispersao_musicas_repetidas_por_artista:.2f}")

# Encontrando o artista com a maior quantidade de músicas repetidas
artista_com_mais_musicas_repetidas = quantidade_musicas_repetidas_por_artista.idxmax()
print(f"Artista com a Maior Quantidade de Músicas Repetidas: {artista_com_mais_musicas_repetidas}")

# Análise Estatística:
print("Análise Estatística:")
print(f"Média de Músicas Repetidas por Artista: {media_musicas_repetidas_por_artista:.2f} músicas repetidas.")
print(f"Interpretação Estatística: A média representa o número médio de músicas repetidas por artista. Em 2023, cada artista teve, em média, 1,76 músicas repetidas, indicando que é comum que artistas tenham, em média, pelo menos uma música repetida.")
print(f"Moda de Músicas Repetidas por Artista: {moda_musicas_repetidas_por_artista} música repetida.")
print(f"Interpretação Estatística: A moda é o valor mais comum, ou seja, 1 música repetida é o valor que mais apareceu. Isso sugere que a maioria dos artistas teve exatamente uma música repetida.")
print(f"Mediana de Músicas Repetidas por Artista: {mediana_musicas_repetidas_por_artista:.2f} música repetida.")
print(f"Interpretação Estatística: A mediana indica que 50% dos artistas tiveram até 1 música repetida, enquanto os outros 50% tiveram mais de 1. A mediana sendo igual à moda reforça que a maioria dos artistas tem uma quantidade pequena de músicas repetidas.")
print(f"Amplitude de Músicas Repetidas por Artista: {amplitude_musicas_repetidas_por_artista} músicas repetidas.")
print(f"Interpretação Estatística: A amplitude, que é a diferença entre o maior e o menor número de músicas repetidas, é de 13 músicas. Isso indica uma grande variação entre os artistas, com alguns artistas repetindo muitas músicas, enquanto outros não repetem tantas.")
print(f"Quartis (Q1 = {Q1_musicas_repetidas_por_artista}, Q3 = {Q3_musicas_repetidas_por_artista}):")
print(f"Interpretação Estatística: O primeiro quartil (Q1) sendo igual a 1 indica que 25% dos artistas têm no máximo 1 música repetida. O terceiro quartil (Q3) de 2 músicas repetidas mostra que 75% dos artistas têm até 2 músicas repetidas. A maior parte dos artistas tem entre 1 e 2 músicas repetidas.")
print(f"Desvio Padrão de Músicas Repetidas por Artista: {desvio_padrao_musicas_repetidas_por_artista:.2f} músicas repetidas.")
print(f"Interpretação Estatística: O desvio padrão relativamente alto indica uma variação considerável no número de músicas repetidas entre os artistas. Isso significa que há uma dispersão notável em torno da média de 1,76 músicas repetidas.")
print(f"Assimetria de Músicas Repetidas por Artista: {assimetria_musicas_repetidas_por_artista}.")
print(f"Interpretação Estatística: A assimetria elevada indica que a distribuição é bastante inclinada para a direita. Isso significa que a maioria dos artistas tem poucas músicas repetidas, mas alguns têm números bem mais altos, puxando a distribuição para o lado positivo.")
print(f"Coeficiente de Dispersão de Músicas Repetidas por Artista: {coeficiente_dispersao_musicas_repetidas_por_artista:.2f}.")
print(f"Interpretação Estatística: O coeficiente de dispersão mede a variação relativa em relação à média. Um valor de 0,89 indica uma dispersão moderada, mostrando que o número de músicas repetidas por artista tem uma variação significativa, embora ainda relativamente concentrada em torno da média.")
print(f"Artista com a Maior Quantidade de Músicas Repetidas: {artista_com_mais_musicas_repetidas}.")
print(f"Interpretação Estatística: O artista com o maior número de músicas repetidas, 'MUSIC LAB JPN', destaca-se significativamente, sugerindo que este artista pode ter uma prática de remixagem ou republicação de suas músicas em maior escala.")


# Criando uma tabela com os resultados
tabela_resultados = pd.DataFrame({
    'Média': [media_musicas_repetidas_por_artista],
    'Moda': [moda_musicas_repetidas_por_artista],
    'Mediana': [mediana_musicas_repetidas_por_artista],
    'Amplitude': [amplitude_musicas_repetidas_por_artista],
    'Q1': [Q1_musicas_repetidas_por_artista],
    'Q3': [Q3_musicas_repetidas_por_artista],
    'Desvio Padrão': [desvio_padrao_musicas_repetidas_por_artista],
    'Assimetria': [assimetria_musicas_repetidas_por_artista],
    'Coeficiente de Dispersão': [coeficiente_dispersao_musicas_repetidas_por_artista],
    'Artista com a Maior Quantidade de Músicas Repetidas': [artista_com_mais_musicas_repetidas]
})

# Exibindo a tabela
salvar_resultados_em_excel(tabela_resultados,r'C:\Users\USER\Desktop\Projeto_Senac\Data_Result\Analise_Musicas Repitidas por Artista de 2014 a 20244\Analise_Musicas Repitidas por Artista de 2014 a 2024.xlsx')

```
